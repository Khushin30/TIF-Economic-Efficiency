```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.cm import get_cmap
import json
from geopy.geocoders import Nominatim
import geopandas as gpd
from shapely.geometry import Point
```

```
tif_geo = gpd.read_file('Boundaries TIF.geojson')
com_geo = gpd.read_file('BoundariesCommunity.geojson')
```

```
com_geo.head()
```

```
import geopandas as gpd

# Assuming you have GeoDataFrames 'tif' and 'com'

# Create an empty dictionary to store the intersections
intersections = {}

# Iterate through each row in 'tif'
for index, tif_row in tif_geo.iterrows():
    tif_geometry = tif_row['geometry']
    intersecting_rows = []
    
    # Check for intersections with each row in 'com'
    for com_index, com_row in com_geo.iterrows():
        com_geometry = com_row['geometry']
        
        # Check if the geometries intersect
        if tif_geometry.intersects(com_geometry):
            intersecting_rows.append(com_row["community"])
    
    # Store the intersecting rows for this tif row
    intersections[tif_row["name"]] = intersecting_rows

# 'intersections' now contains a dictionary where keys are tif row indices
# and values are lists of com row indices that intersect with each tif row
```

```
intersections
```

```
with open('intersections.json', 'w') as f:
    json.dump(intersections, f)
```

```
# to load to dictionary
with open('intersections.json', 'r') as f:
    intersections = json.load(f)
```
